# Notes
- In the turboquant paper, the embeddings are rotated such that each coordinate of the $d$-dimensional vector is nearly independent from each other and follows a symmetric beta distribution
- The Beta distribution that is used is such that $B(\alpha, \beta)(u)$ with $u\in[0, 1]$, where $\alpha=\beta=(d-1) / 2$, and the random variable is shift by 1 and scaled by 2 ($x=2u-1$)

```python
from scipy.stats import beta
from scipy.cluster.vq import kmeans
```

```python
d = 384 # embedding dimension 
bits = 2 # number of bits per dimension
```

```python
a = b = (d-1) / 2
```

```python
samples = beta.rvs(a, b, loc=-1, scale=2, size=int(1e6)) #shifted by 1, scaled by 2
```

```python
samples
```

```python
centroids, _ = kmeans(samples, int(2**bits))
```

```python
centroids.sort()
```
